```python
from pyspark.sql.types import *
```

## Read Data From Blob Storage

```python
containerName = ""
filePath = ""
storageAccount = ""
storageKey = ""

# End point and Conn String
storageEndpoint = "wasbs://"+containerName + "@" + storageAccount + ".blob.core.windows.net"
storageConnSting = "fs.azure.account.key." + storageAccount + ".blob.core.windows.net"

spark.conf.set(storageConnSting, storageKey)
storageEndpointFolder = storageEndpoint + "/" + filePath
```

```python
display(dbutils.fs.ls(storageEndpointFolder))
```

```python
# Create a Schema
customer_schema = StructType([
    StructField("C_CUSTKEY", IntegerType()),
    StructField("C_NAME", StringType()),
    StructField("C_ADDRESS", StringType()),
    StructField("C_NATIONKEY", ShortType()),
    StructField("C_PHONE", StringType()),
    StructField("C_ACCTBAL", DecimalType(18,2)),
    StructField("C_MKTSEGMENT", StringType()),
    StructField("C_COMMENT", StringType())
])

# read the data

df_customer = spark.read.format("csv").option("header", True).schema(customer_schema).load(storageEndpointFolder+"/part-0*.csv")

display(df_customer.limit(10))
```

## Writing & Reading Data To Azure SQL DB through JDBC Driver

- Java Database Connectivity

```python
# server and Database details
server_name = ""
server_user_name = ""
server_password = ""
database_name = ""
table_name = ""

# create jdbc URL
jdbcHostname = "{}.database.windows.net".format(server_name)
jdbcDatabase = database_name
jdbcPort = 1433
jdbcUrl = "jdbc:sqlserver://{0}:{1};database={2}".format(jdbcHostname, jdbcPort, jdbcDatabase)

# Connection properties
connectionProperties = {
"user" : server_user_name,
"password" : server_password,
"driver" : "com.microsoft.sqlserver.jdbc.SQLServerDriver"
}

# read teh table
Spdf = spark.read.jdbc(url=jdbcUrl, table=table_name, properties=connectionProperties)
```

```python
display(Spdf.limit(5))
```

```python
# writing data to the table 

df_customer.write.jdbc(
    url=jdbcUrl, 
    mode = "append",
    table=table_name, 
    properties = connectionProperties
)
```

```python
# read the data from Azure SQL
df_jdbc_read = spark.read.jdbc(
    url = jdbcUrl,
    table = table_name,
    properties = connectionProperties
)

df_jdbc_read.show(5)
```
