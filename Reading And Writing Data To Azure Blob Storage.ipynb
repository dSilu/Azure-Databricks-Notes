## Writing Data to Azure Blob Storage

```python
# mount the folder
storageAccount = ""
containerName = ""
storageKey = ""
mountpoint = "/mnt/Blob"

storageEndpoint = "wasbs://{}@{}.blob.core.windows.net".format(containerName, storageAccount)
storageConnSting = "fs.azure.account.key.{}.blob.core.windows.net".format(storageAccount)

# destination
parquetCustomerDestMount = "{}/Customer.parquetFiles".format(mountpoint)

try:
    dbutils.fs.mount(
        source = storageEndpoint,
        mount_point = mountpoint,
        extra_configs = {
            storageConnSting: storageKey
        }
    )
except Exception as e:
    print("Error \n", e)
```

```python
%fs ls /mnt/Blob
```

```python
display(dbutils.fs.ls("/mnt/Blob/customers/"))
```

```python
# reading the files to a dataframe

df_customer = spark.read.format("csv").option("header", True).load("/mnt/Blob/customers")
```

```python
df_customer.printSchema()
```

```python
df_customer.show(5)
```

```python
# print 10 records from the dataframe
display(df_customer.limit(10))
```

```python
# writing data to Azure Blob
df_customer_partitioned = df_customer.repartition(10)
df_customer_partitioned.write.mode("overwrite").option("header", True).parquet(parquetCustomerDestMount)
```

## Reading Data from Azure Blob Storage Using Dataframe API Without Mount Point

```python
folderPath = ""
containerName = ""
storageAccount = ""
storageKey = ""

storageEndpoint = "wasbs://{}@{}.blob.core.windows.net".format(containerName, storageAccount)
storageConnSting = "fs.azure.account.key.{}.blob.core.windows.net".format(storageAccount)

# set account access key 

spark.conf.set(storageConnSting, storageKey)

storageEndpointFolders = "{}/{}".format(storageEndpoint, folderPath)
```

```python
# reading files
display(dbutils.fs.ls(storageEndpointFolders))
```

```python
# define a schema
from pyspark.sql.types import *

customer_schema = StructType([ 
    StructField("CustomerKey", IntegerType()),
    StructField("Name", StringType()),
    StructField("Address", StringType()),
    StructField("Nationkey", ShortType()),
    StructField("Phone", StringType()),
    StructField("ACCTBAL", DoubleType()),
    StructField("MKTSegment", StringType()),
    StructField("Comment", StringType())
])
```

```python
# reading files directly from the Azure Blob storage using wasbs path

df_customer = spark.read.format("csv").option("header", True).schema(customer_schema).\
    load("wasbs://{}@{}.blob.core.windows.net/{}/".format(containerName, storageAccount, folderPath))
```

```python
df_customer.count()
```

## Writing Data Directly to Azure Blob without Mount Point

```python
destination_directory = "wasbs://{}@{}.blob.core.windows.net/Writingfolder".format(containerName, storageAccount)
df_customer_partitioned = df_customer.repartition(15)
df_customer_partitioned.write.mode("overwrite").option("header", True).csv(destination_directory)
```

```python
# checking directory
storageEndpoint = "wasbs://{}@{}.blob.core.windows.net".format(containerName, storageAccount)
storageEndpointFolders = storageEndpoint + "/Writingfolder"
display(dbutils.fs.ls(storageEndpointFolders))
```

## Access Data From Azure Blob Storage with SAS

- SAS stands for Shared Access Signature

```python
containerName = ""
storageAccount = ""
sasToken = ""

storageConnSting = "fs.azure.sas" + containerName + storageAccount + ".blob.core.windows.net"

spark.conf.set(storageConnSting, sasToken)
```

```python
# establish storage point
storageEndpoint = "wasbs://"+containerName + "@" + storageAccount + ".blob.core.windows.net"
storageEndpointFolder = storageEndpoint + "/Writingfolder"

display(dbutils.fs.ls(storageEndpointFolder))
```

```python
df = spark.read.format("csv").option("header", True).load(storageEndpointFolder+"/part-0*.csv")
display(df.limit(5))
```
